```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scienceplots
# plt.style.use(['science', 'grid'])
plt.style.use(['science'])
```

```python
fig, axs = plt.subplots(2, 1, figsize=(7.5, 4), sharex=True, sharey=True, dpi=300)

ax = axs[0]
x = np.load('../../data/bin_10ms/A1.npy', allow_pickle=True)
# evts = np.concatenate(x[:10])
x = [e for e in x if e['t'].shape[0] > 5e3]
evts = x[1]
n_bins = 1_000
data = np.histogram(evts['t'] - evts['t'].min(), bins=n_bins)
ax.set_title('Particles measured for the present experiment')
ax.bar(data[1][:-1] / 1e4, data[0] / data[0].max(), width=(data[1][1]-data[1][0])/1e4)

ax = axs[1]
x = np.load('Event_count_YAR.npy')
ax.set_title('Biological cells measured for a forthcoming experiment')
ax.bar(np.arange(0, 1, 1./len(x)), x/x.max(), width=1/len(x))

axs[1].set_xlabel('Time (normalized)')
for ax in axs:
    # ax.grid(False)
    # ax.set_ylabel('Normalized event count')
    # ax.set_xlabel('Time (normalized)')
    ax.set_ylim(-0.05, 1.05)

plt.tight_layout()
plt.savefig('event_count_viz_.pdf')
```
